```python
from pathlib import Path
import polars as pl # 使用 polars

# cwd = Path(../Downloads/)# 檔案路徑
raw = pl.read_excel(
    cwd / "material list.xlsx",
    read_options={"header_row": 1} # 跳過第一行
)

ok_df = raw.with_columns(
    pl.col('Fab1').fill_null(strategy="forward"),
    pl.col('Phase1').fill_null(strategy="forward"),
    pl.col('Vendor').fill_null(strategy="forward"),
    pl.col('Vendor Area(Country)').fill_null(strategy="forward"),
    pl.col('Material manufacture(country)').fill_null(strategy="forward"),
    pl.col('release BU').fill_null(strategy="forward"),
) # 補齊各欄位
ok_df.write_excel(cwd / "material_OK.xlsx")
```
